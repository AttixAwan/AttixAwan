```python
# --- STEP 1: Imports and Data Preparation ---

import os
import numpy as np
import pandas as pd
from glob import glob
import matplotlib.pyplot as plt
import seaborn as sns

# Folder containing your stock .xlsx files
folder = 'Stock_Files_2024'
stock_data = {}
all_lengths = []

for path in glob(os.path.join(folder, '*.xlsx')):
    symbol = os.path.basename(path).split('.')[0]
    try:
        df = pd.read_excel(path)[['Date', 'OPEN_PRICE', 'CLOSING_PRICE', 'HIGH', 'LOW', 'TRADED_VOLUME']]
        df['Date'] = pd.to_datetime(df['Date'])
        df = df.sort_values('Date')
        df.set_index('Date', inplace=True)
        weekly_df = df.resample('W-FRI').agg({
            'OPEN_PRICE': 'first',
            'CLOSING_PRICE': 'last',
            'HIGH': 'max',
            'LOW': 'min',
            'TRADED_VOLUME': 'sum'
        }).dropna()
        weekly_df['return'] = weekly_df['CLOSING_PRICE'].pct_change()
        weekly_df = weekly_df.dropna().reset_index()
        stock_data[symbol] = weekly_df
        all_lengths.append((symbol, len(weekly_df)))
    except Exception as e:
        print(f"Error reading {symbol}: {e}")

print(f"‚úÖ Total stocks loaded: {len(stock_data)}")
print("üìä Weekly data lengths (top 5):")
print(sorted(all_lengths, key=lambda x: x[1], reverse=True)[:5])

# Show sample data
first_symbol = list(stock_data.keys())[0]
print(f"\nüîç Weekly data sample for {first_symbol}:\n")
print(stock_data[first_symbol].head())
```

```python
# --- STEP 2: Feature Matrix Construction & EDA ---

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Fix common sequence length for all stocks (e.g., last 52 weeks)
common_length = 52
symbols = []
feature_matrix = []
return_matrix = []

feature_cols = ['OPEN_PRICE', 'CLOSING_PRICE', 'HIGH', 'LOW', 'TRADED_VOLUME']

for symbol, df in stock_data.items():
    if len(df) >= common_length:
        df = df.tail(common_length).copy()
        features = df[feature_cols].values
        returns = df['return'].values
        if not np.any(np.isnan(features)) and not np.any(np.isnan(returns)):
            symbols.append(symbol)
            feature_matrix.append(features)
            return_matrix.append(returns)

feature_matrix = np.stack(feature_matrix)
return_matrix = np.stack(return_matrix)

print("Feature matrix shape:", feature_matrix.shape)  # (num_stocks, 52, 5)
print("Return matrix shape :", return_matrix.shape)   # (num_stocks, 52)
print("Symbols (sample):", symbols[:5])
print("Feature sample (first stock, first week):", feature_matrix[0, 0, :])
print("Return sample  (first stock):", return_matrix[0, :5])

# --- EDA Visuals ---

# 1. Distribution of weekly returns (all stocks, all weeks)
plt.figure(figsize=(8,4))
all_returns = return_matrix.flatten()
plt.hist(all_returns, bins=50, color='skyblue', edgecolor='black')
plt.title("Distribution of Weekly Returns (All Stocks)")
plt.xlabel("Weekly Return")
plt.ylabel("Frequency")
plt.show()

# 2. Correlation heatmap (returns, last 52 weeks)
corr_df = pd.DataFrame(return_matrix.T, columns=symbols)
corr_matrix = corr_df.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, cmap='coolwarm', center=0)
plt.title("Return Correlation Heatmap (Across Stocks)")
plt.show()
```

```python
# --- Full-batch GAT+BiLSTM (no DataLoader batching) ---

import torch
import torch.nn as nn
from torch_geometric.nn import GATConv

# GAT + BiLSTM Model
class GAT_BiLSTM(nn.Module):
    def __init__(self, input_dim, hidden_dim, gat_heads, lstm_hidden, num_weeks):
        super().__init__()
        self.gat = GATConv(input_dim, hidden_dim, heads=gat_heads, concat=True)
        self.bilstm = nn.LSTM(hidden_dim * gat_heads, lstm_hidden, bidirectional=True, batch_first=True)
        self.fc = nn.Linear(lstm_hidden * 2, num_weeks)
    def forward(self, x, edge_index):
        # x: [num_stocks, weeks, features]
        num_stocks, weeks, features = x.shape
        # Apply GAT to each week across all stocks
        out = []
        for t in range(weeks):
            xt = x[:, t, :]  # [num_stocks, features]
            xt = self.gat(xt, edge_index)
            out.append(xt)
        x_gat = torch.stack(out, dim=1)  # [num_stocks, weeks, gat_hidden*heads]
        lstm_out, _ = self.bilstm(x_gat)
        out = self.fc(lstm_out[:, -1, :])  # final output: [num_stocks, num_weeks]
        return out

# Prepare everything for full-batch
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
X_all = torch.tensor(feature_matrix, dtype=torch.float32).to(device)
y_all = torch.tensor(return_matrix, dtype=torch.float32).to(device)

# Pearson-correlation edge_index for the full stock graph
corr = np.corrcoef(return_matrix)
threshold = 0.3
edge_index = np.array(np.where(np.abs(corr) > threshold))
edge_index = edge_index[:, edge_index[0] != edge_index[1]]  # remove self-loops
edge_index_torch = torch.tensor(edge_index, dtype=torch.long).to(device)

# Train/Test split indices
split = int(0.8 * X_all.shape[0])
train_idx = np.arange(split)
test_idx = np.arange(split, X_all.shape[0])

# Model, loss, optimizer
model = GAT_BiLSTM(input_dim=5, hidden_dim=16, gat_heads=2, lstm_hidden=32, num_weeks=52).to(device)
loss_fn = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

# --- Full-batch training ---
epochs = 20
for epoch in range(epochs):
    model.train()
    optimizer.zero_grad()
    pred = model(X_all, edge_index_torch)
    loss = loss_fn(pred[train_idx], y_all[train_idx])
    loss.backward()
    optimizer.step()
    print(f"Epoch {epoch+1}/{epochs} - Train Loss: {loss.item():.4f}")

# --- Predict for all stocks ---
model.eval()
with torch.no_grad():
    pred_all = model(X_all, edge_index_torch).cpu().numpy()
    predicted_last_return = pred_all[:, -1]

top_n = 10
top_indices = np.argsort(predicted_last_return)[-top_n:][::-1]
top_symbols = [symbols[i] for i in top_indices]
print("\nTop stocks by predicted return:", top_symbols)
print("Predicted returns:", predicted_last_return[top_indices])
```

```python
# --- STEP 4: Portfolio Construction, Metrics, Visuals ---

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Get historical returns for your selected top stocks
selected_returns = []
selected_symbols = top_symbols  # from previous cell
for symbol in selected_symbols:
    idx = symbols.index(symbol)
    selected_returns.append(return_matrix[idx])
selected_returns = np.array(selected_returns)  # shape: (top_n, 52)

# 1. Equal-weight portfolio returns
portfolio_ret = selected_returns.mean(axis=0)  # [52,]

# 2. Cumulative return
cum_ret = (1 + portfolio_ret).cumprod()

# 3. Drawdown calculation
cum_max = np.maximum.accumulate(cum_ret)
drawdown = (cum_ret - cum_max) / cum_max

# 4. Metrics
rf = 0.0  # risk-free rate
ann_factor = 52  # weekly to annual

ann_ret = (cum_ret[-1])**(ann_factor/len(cum_ret)) - 1
ann_vol = np.std(portfolio_ret) * np.sqrt(ann_factor)
sharpe = (ann_ret - rf) / ann_vol
max_dd = drawdown.min()

print(f"Annualized Return: {ann_ret:.2%}")
print(f"Annualized Volatility: {ann_vol:.2%}")
print(f"Sharpe Ratio: {sharpe:.2f}")
print(f"Maximum Drawdown: {max_dd:.2%}")

# 5. Plot cumulative return
plt.figure(figsize=(8,4))
plt.plot(cum_ret, label='Portfolio')
plt.title("Portfolio Cumulative Return")
plt.xlabel("Week")
plt.ylabel("Cumulative Return")
plt.legend()
plt.show()

# 6. Plot drawdown
plt.figure(figsize=(8,4))
plt.plot(drawdown, color='red')
plt.title("Portfolio Drawdown")
plt.xlabel("Week")
plt.ylabel("Drawdown")
plt.show()

# 7. Heatmap of correlation for selected stocks
plt.figure(figsize=(7,6))
df_sel = pd.DataFrame(selected_returns.T, columns=selected_symbols)
sns.heatmap(df_sel.corr(), annot=True, cmap='coolwarm', center=0)
plt.title("Return Correlation (Top Stocks)")
plt.show()
```

```python
import numpy as np

def downside_risk(returns, rf=0.0, freq=52):
    downside = np.where(returns < rf, returns, 0)
    dr = np.sqrt((downside**2).sum() / len(returns)) * np.sqrt(freq)
    return dr

def sortino_ratio(returns, rf=0.0, freq=52):
    ann_ret = (np.prod(1+returns))**(freq/len(returns)) - 1
    dr = downside_risk(returns, rf, freq)
    if dr == 0:
        return np.nan
    return (ann_ret - rf) / dr

def calmar_ratio(returns, freq=52):
    cum_ret = (1 + returns).cumprod()
    annual_return = cum_ret[-1]**(freq/len(cum_ret)) - 1
    running_max = np.maximum.accumulate(cum_ret)
    max_dd = np.min((cum_ret - running_max) / running_max)
    if max_dd == 0:
        return np.nan
    return annual_return / abs(max_dd)

def semideviation(returns):
    mean = np.mean(returns)
    negative = returns[returns < mean]
    return np.std(negative, ddof=1)

def value_at_risk(returns, alpha=0.05):
    # Historical VaR
    return np.percentile(returns, 100 * alpha)

def conditional_value_at_risk(returns, alpha=0.05):
    var = value_at_risk(returns, alpha)
    return returns[returns <= var].mean()

# --- Try alternative weighting: volatility-adjusted ("risk parity lite") ---
equal_weights = np.ones(selected_returns.shape[0]) / selected_returns.shape[0]
inv_vol = 1 / np.std(selected_returns, axis=1)
risk_parity_weights = inv_vol / np.sum(inv_vol)

# Choose weighting scheme: 'equal' or 'risk_parity'
weighting = 'equal'  # change to 'risk_parity' to test

if weighting == 'equal':
    weights = equal_weights
    print("Using equal weighting.")
else:
    weights = risk_parity_weights
    print("Using volatility-adjusted (risk parity) weighting.")

# Weighted portfolio returns
portfolio_ret_w = np.dot(weights, selected_returns)

# Advanced metrics
sharpe = (np.prod(1+portfolio_ret_w))**(ann_factor/len(portfolio_ret_w)) - 1
sharpe = (sharpe - rf) / (np.std(portfolio_ret_w) * np.sqrt(ann_factor))
sortino = sortino_ratio(portfolio_ret_w)
calmar = calmar_ratio(portfolio_ret_w)
semi_dev = semideviation(portfolio_ret_w)
downside = downside_risk(portfolio_ret_w)
var_95 = value_at_risk(portfolio_ret_w, 0.05)
cvar_95 = conditional_value_at_risk(portfolio_ret_w, 0.05)

print(f"\nSharpe Ratio: {sharpe:.2f}")
print(f"Sortino Ratio: {sortino:.2f}")
print(f"Calmar Ratio: {calmar:.2f}")
print(f"Semi-deviation: {semi_dev:.4f}")
print(f"Downside Risk (annualized): {downside:.4f}")
print(f"Value at Risk (5%%): {var_95:.4f}")
print(f"Conditional Value at Risk (CVaR 5%%): {cvar_95:.4f}")

# Plot: Portfolio returns (equal vs. risk parity)
import matplotlib.pyplot as plt

plt.figure(figsize=(8,4))
plt.plot((1+portfolio_ret).cumprod(), label='Equal Weight')
plt.plot((1+np.dot(risk_parity_weights, selected_returns)).cumprod(), label='Risk Parity Weight', linestyle='--')
plt.title("Portfolio Cumulative Return: Equal vs. Risk Parity")
plt.xlabel("Week")
plt.ylabel("Cumulative Return")
plt.legend()
plt.show()
```

```python
import numpy as np

# Correlation matrix among selected stocks
corr_matrix = np.corrcoef(selected_returns)
np.fill_diagonal(corr_matrix, 0)  # Ignore self-correlation

# Greedy selection: start with the very top stock, then add stocks with lowest average correlation to those already picked.
diversified_indices = [0]  # Start with highest predicted return
n_diverse = 10  # Set size of diversified portfolio (<= len(top_symbols))

while len(diversified_indices) < n_diverse:
    not_picked = [i for i in range(len(top_symbols)) if i not in diversified_indices]
    min_corrs = []
    for i in not_picked:
        avg_corr = np.mean(np.abs(corr_matrix[i, diversified_indices]))
        min_corrs.append(avg_corr)
    next_idx = not_picked[np.argmin(min_corrs)]
    diversified_indices.append(next_idx)

diversified_symbols = [top_symbols[i] for i in diversified_indices]
print("Diversified Portfolio Symbols:", diversified_symbols)

# Get returns for diversified portfolio
diverse_returns = selected_returns[diversified_indices]
diverse_portfolio_ret = diverse_returns.mean(axis=0)
diverse_cum_ret = (1 + diverse_portfolio_ret).cumprod()
```
